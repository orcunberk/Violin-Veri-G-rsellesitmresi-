```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
df = pd.read_csv(url, sep = "|")
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
sns.violinplot(data = df, x = "gender", y ="age");
```

```python
width = 15
height = 10
plt.figure(figsize = (width,height))
ax = sns.violinplot(data = df, x = "gender",y = "age", linewidth = 5, width = 0.8, palette = "cool")
plt.title("Cinsiyete Göre Yaş Dağılımları")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
#Keman grafiği her grubu için gözlem sayısını belirlemek için iyi bir uygulamadır.
```

```python
width = 10
height = 6
plt.figure(figsize = (width, height))
ax = sns.violinplot(data = df, x = "gender", y = "age", linewidth = 3, width = 0.5, palette = "cool", inner = "stick")
plt.title("Cinsiyete Göre Yaş Dağılımı")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```
